```python
#!/usr/bin/python
 
import math
import matplotlib.pyplot as plt
LOT_SIZE = 80
POSITION_LIMIT = 100
SIGMOID_C = math.ceil(math.log(LOT_SIZE-1)/-0.7)

def sigmoid(x):     # Author: Yip
	return 1/(1+math.exp(-SIGMOID_C*(x)))

def get_num(position):
	return int(sigmoid(position/(2*POSITION_LIMIT))*LOT_SIZE)

order_num = list()
position_num = list()

for i in range(-99,100):
	position_num.append(i)
	order_num.append(get_num(i))
	
plt.plot(position_num,order_num)
plt.show()
```
